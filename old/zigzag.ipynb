```python

from sklearn.discriminant_analysis import StandardScaler


df = pd.read_csv('data/NASDAQ.csv')
#print(df.head())  # 7 columns, including the Date.

cols = list(df)[1:5]
# New dataframe with only training data - 5 columns
df_for_training = df[cols].astype(float)

df_for_training = df_for_training.tail(200)
```

```python

# LSTM uses sigmoid and tanh that are sensitive to magnitude so values need to be normalized
# normalize the dataset
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
data = scaler.transform(df_for_training)
print(data)
```

```python

num_samples = len(data)
sequence_length = 20
```

```python
# Create synthetic zigzag labels
zigzag_indices = np.arange(0, num_samples, 5)
labels = np.zeros(num_samples)
labels[zigzag_indices] = 1

# Plot the data and zigzag patterns
fig, axs = plt.subplots(nrows=num_samples, sharex=True, figsize=(28, 12))
for i in range(num_samples):
    candle_data = data[i]
    ax = axs[i]

    # Plot candlestick
    candle_colors = ['g' if candle[3] >= candle[0]
                     else 'r' for candle in candle_data]
    ax.vlines(np.arange(sequence_length),
              candle_data[:, 1], candle_data[:, 2], color=candle_colors, linewidth=2)
    ax.plot(np.arange(sequence_length), candle_data[:, 0], 'k-', linewidth=2)

    # Highlight zigzag patterns
    if labels[i] == 1:
        ax.axvline(x=sequence_length - 1, color='b', linestyle='--')

    ax.set_xlim(0, sequence_length - 1)
    ax.set_ylim(min(candle_data[:, 2]) - 0.1, max(candle_data[:, 1]) + 0.1)
    ax.set_yticks([])
    ax.set_xticks([])
```

```python

plt.tight_layout()
plt.show()
```
